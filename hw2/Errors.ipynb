# Ошибки в классификации

Итак, у нас есть несколько моделей, которые мы обучали классифицированию текстов. Данные были обработаны по-разному, и сказать о данных можно следующее:
1. Пунктуация -- довольно сильный признак, но -- в основном благодаря скобочкам, которые в русскоязычном сегменте интернета используются как смайлики. А поскольку сам датасет когда-то был классифицирован по смайликам, вероятно, получается некоторая очень сильная фича для этого датасета, которая не очень сработает, например, на английском.
2. Лемматизированный корпус лучше, чем не лемматизированный. Звучит очевидно, конечно, но работет.<br>
 <br>
Далее, наши модели:<br>
1. И обученные с пунктуацией, и обученные без неё модели склонны переобучаться.
2. Уменьшение количества эпох, лёргнинг-рейта и введение дропаута помогает.

## Метрики:

Accuracy, precision и recall выглядят на тесте почти идеально на моделях, обученных с пунктуацией. Но при этом и у них есть места, где они ошибаются. Например, 'rt : смешной' был классифицирован как моделью как положительный, хотя в датасете таким не отмечен. Места с иронией и срказмом в целом плохо поддаются такой обработке
